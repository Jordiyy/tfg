```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('score_manual.csv', sep=';')
target_score = data['score']
data_low_col = pd.read_csv('score_low_col.csv', sep=',')
target_low_col_score = data_low_col['score']
data_low_systolic = pd.read_csv('score_low_systolic.csv', sep=',')
target_low_systolic_score = data_low_systolic['score']
```

```python
fig, ax = plt.subplots()

# Dibujar el histograma
ax.hist(target_score, bins=10, color='orange', alpha=0.7, edgecolor='k')

# Etiquetar ejes y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de valores SCORE')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)

# Dibujar el histograma
ax1.hist(target_low_col_score, bins=10, color='orange', alpha=0.7, edgecolor='k')
ax2.hist(target_low_col_score[target_low_col_score < 20], bins=10, color='orange', alpha=0.7, edgecolor='k')
ax2.set_title("SCORE < 20", fontsize=8)

# Etiquetar ejes y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.suptitle('Histograma valores SCORE con baja incidencia de colesterol')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)

# Dibujar el histograma
ax1.hist(target_low_systolic_score, bins=10, color='orange', alpha=0.7, edgecolor='k')
ax2.hist(target_low_systolic_score[target_low_systolic_score < 20], bins=10, color='orange', alpha=0.7, edgecolor='k')
ax2.set_title("SCORE < 20", fontsize=8)

# Etiquetar ejes y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.suptitle('Histograma valores SCORE con baja incidencia de presión sistólica')
```
